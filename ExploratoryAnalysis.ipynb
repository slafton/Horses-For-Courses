# runners.csv

```python
# Import and read the runners.csv
runners_df = pd.read_csv("resources/runners.csv")
runners_df.head()
```

```python
# remove excess columns
runners_df.drop(['tip_nine_place', 'tip_nine_win', 'tip_eight_place', 'tip_eight_win', 'tip_seven_place', 'tip_seven_win', 'tip_six_win', 'tip_six_place', 'tip_five_win', 'tip_five_place', 'tip_four_win', 'tip_four_place', 'tip_three_win', 'tip_three_place', 'tip_two_place', 'tip_two_win', 'tip_one_place', 'tip_one_win'],axis=1, inplace=True)

runners_df.head()
```

```python
runners_df.drop(['collected_at', 'market_id', 'position', 'place_paid', 'margin', 'blinkers', 'emergency', 'form_rating_one', 'form_rating_two', 'form_rating_three'],axis=1, inplace=True)
runners_df.head()
```

```python
runners_df.drop(['barrier', 'number', 'last_five_starts', 'favourite_odds_win', 'favourite_odds_place', 'favourite_pool_place', 'favourite_pool_win'],axis=1, inplace=True)
runners_df.head()
```

```python
runners_df.drop(['horse_id'],axis=1, inplace=True, )
runners_df.head()
```
